```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    f1_score,
    roc_auc_score
)
df = pd.read_csv(r"D:\college\appliedmaths\Teen_Mental_Health_Dataset.csv")

# Features and target
X = df.drop(columns=['depression_label'])
y = df['depression_label']

# Convert text columns into numbers
X = pd.get_dummies(X, drop_first=True, dtype=int)

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=1,
    stratify=y
)

# Decision Tree model
dt = DecisionTreeClassifier(random_state=1)

# Train model
dt.fit(X_train, y_train)

# Prediction
y_pred = dt.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))

print("F1 Score:", f1_score(y_test, y_pred))

print("ROC-AUC:", roc_auc_score(y_test, y_pred))

print(classification_report(y_test, y_pred))
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
grid = GridSearchCV(
    DecisionTreeClassifier(random_state=42),
    param_grid,
    cv=5,
    scoring='f1',
    verbose=2,
    n_jobs=-1
)
grid.fit(X_train, y_train)
print("Best Parameters:")
print(grid.best_params_)
best_dt = grid.best_estimator_

y_pred_best = best_dt.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred_best))

print("F1 Score:", f1_score(y_test, y_pred_best))

print("ROC-AUC:", roc_auc_score(y_test, y_pred_best))

print(classification_report(y_test, y_pred_best))
cm = confusion_matrix(y_test, y_pred_best)

print("Confusion Matrix:")
print(cm)



```

```python
X_reduced = X.drop(columns=[
    'platform_usage_Instagram',
    'platform_usage_TikTok',
    'gender_male'
])

X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(
    X_reduced,
    y,
    test_size=0.2,
    random_state=1,
    stratify=y
)
best_dt.fit(X_train_r, y_train_r)
y_pred_r = best_dt.predict(X_test_r)

print("Reduced Feature Model Results")

print("Accuracy after removing features:", accuracy_score(y_test_r, y_pred_r))

print("F1 Score after removing features:", f1_score(y_test_r, y_pred_r))

print("ROC-AUC after removing features:", roc_auc_score(y_test_r, y_pred_r))

print(classification_report(y_test_r, y_pred_r))
cm = confusion_matrix(y_test_r, y_pred_r)

print("Confusion Matrix:")
print(cm)
```
